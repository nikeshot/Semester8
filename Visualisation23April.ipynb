```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
data = pd.read_excel("C:/Users/Niku/Documents/Semester8/DataFiles/CommViolPredUnnormalizedData.xlsx")
data.shape
```

```python
rows = data.shape[0]
cols = data.shape[1]
count = np.zeros(cols)
cnt = 0 
listcols = []
listrows = []
```

```python
data = data.replace('?', np.nan)
missing = data.isnull().sum()
missing = missing*1.0/rows

for i in range(cols):
    if missing[i] < 0.5:
        listcols.append(i)

Viol = data['ViolentCrimesPerPop']

for i in range(rows):
    if ~np.isnan(Viol[i]):
        listrows.append(i)

datanew = data.iloc[listrows, listcols]
datanew.shape
```

```python
datanew.fillna(datanew.mean(), inplace=True)
```

```python
datanew = datanew.reset_index()
```

```python
indx = datanew.iloc[0].index
Viol = datanew['ViolentCrimesPerPop']
outcolnum = indx.get_loc('ViolentCrimesPerPop')
Viol = np.array(Viol)
mx = max(Viol)
mn = min(Viol)
rows = datanew.shape[0]
cols = datanew.shape[1]
for i in range(rows):
    Viol[i] = 1.0*(Viol[i]-mn)/(mx-mn)
```

```python
cnt = 0
Category = np.zeros(rows)
for i in range(rows):
    if Viol[i] > 0.1:
        Category[i] = 1
        cnt += 1
print cnt, 1 - cnt*1.0/rows
```

```python
datanew['Category'] = Category
```

```python
del datanew['ViolentCrimesPerPop']
del datanew['fold']
del datanew['communityname']
del datanew['index']
```

```python
sorted_states = sorted(np.unique(datanew['state']))
en_states = enumerate(sorted_states)
list_states = list(en_states)
dict_states = {name: i for i, name in list_states}
datanew.state = datanew.state.map(lambda x: dict_states[x]).astype(int)
```

```python
writer = pd.ExcelWriter('C:/Users/Niku/Documents/Semester8/DataFiles/dataset1.xlsx', engine='xlsxwriter')
datanew.to_excel(writer,'Sheet1')
writer.save()
```

```python
datanew1 = np.matrix(datanew)
rows, cols = datanew1.shape
for i in range(1, cols-1):
    mx = (np.array(max(datanew1[:, i]))).ravel()
    mn = (np.array(min(datanew1[:, i]))).ravel()
    for j in range(rows):
        datanew1[j, i] = 1.0*(datanew1[j, i] - mn)/(mx-mn)
```

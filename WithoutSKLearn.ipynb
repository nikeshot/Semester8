```python
import pandas as pd
import numpy as np
import gzip,csv
from sklearn.model_selection import train_test_split
from collections import Counter, defaultdict
data = pd.read_excel("C:\Users\Niku\Documents\ML Project\data5k.xlsx")
X = data[['DayOfWeek', 'PdDistrict']]
y = data['Category']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.5, random_state=1)
train = pd.concat((X_train, y_train), axis=1)
```

```python
N = float(len(train))
DaysOfWeek = sorted(np.unique(train['DayOfWeek']))
PdDistricts = sorted(np.unique(train['PdDistrict']))
classes = sorted(np.unique(train['Category']))
Sol =['Id']
Sol += classes
trainS = train[['DayOfWeek','PdDistrict','Category']]
```

```python
N = float(len(train))
DaysOfWeek = sorted(np.unique(train['DayOfWeek']))
PdDistricts = sorted(np.unique(train['PdDistrict']))
classes = sorted(np.unique(train['Category']))
Sol =['Id']
Sol += classes
trainS = train[['DayOfWeek','PdDistrict','Category']]

dictDay = {}
dictCat = {}
dictDayWithCat={}

for i in DaysOfWeek:
    for j in classes:
        selected = trainS[ trainS['Category']==j][['DayOfWeek']]
        selectedDay = len(trainS[trainS['DayOfWeek']==i][['DayOfWeek']])
        if len(selected)!=0:
            selectedWithDay = selected [ selected['DayOfWeek']==i]['DayOfWeek']
            probDbyC = len(selectedWithDay)/float(len(selected))
            probD = selectedDay/N
            probC = len(selected)/N
            dictDay[i]=probD
            dictCat[j]=probC
            dictDayWithCat[(i,j)]=probDbyC

dictDis = {}
dictDisWithCat={}

for i in PdDistricts:
    for j in classes:
        selected = trainS[ trainS['Category']==j][['PdDistrict']]
        selectedDis = len(trainS[trainS['PdDistrict']==i][['PdDistrict']])
        if len(selected)!=0:
            selectedWithDis = selected [ selected['PdDistrict']==i]['PdDistrict']
            probDbyC = len(selectedWithDis)/float(len(selected))
            probD = selectedDis/N
            dictDis[i]=probD
            dictDisWithCat[(i,j)]=probDbyC
```

```python
test = X_test
dictCat
```

```python
def maximum(dict1):
    values = dict1.values()
    keys = dict1.keys()
    max_index = values.index(max(values))
    max_key = keys[max_index]
    return max_key

predicted_y = []
ids = 0
datafr = pd.DataFrame(columns=classes)
for (dist,day) in zip(test['PdDistrict'],test['DayOfWeek']):
    predict_dict = {}
    for _class in classes:
        ans1 = dictDisWithCat[(dist,_class)] * dictDayWithCat[(day,_class)] * dictCat[_class]
        print ans1
        ans2 = dictDay[day] * dictDis[dist]
        ans = ans1/ans2
        predict_dict[_class] = ans
    datafr.loc[ids] = predict_dict
    print "Next"
    ids += 1
    p = maximum(predict_dict)
    predicted_y.append(p)    
```

```python
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, classification_report, confusion_matrix
y_pred = np.array(predicted_y)
print y_pred.shape
accuracy_score(y_test, y_pred)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import log_loss
log_loss(y_test, datafr, labels = classes)
```

```python
y_test_series = pd.Series(y_test, name = 'Actual')
y_pred_series = pd.Series(y_pred, name = "Predicted")
df_confusion = pd.crosstab(y_test_series, y_pred_series, margins=True)
df_conf_norm = df_confusion / df_confusion.sum(axis=1)

writer = pd.ExcelWriter('PythonExport1.xlsx')
df_confusion.to_excel(writer,'Sheet1')
writer.save()
```
